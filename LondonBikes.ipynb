```python
new_column={
    'timestamp':'time',
    'cnt':'count',
    't1':'temperature_real_C',
    't2':'temperature_feels_like_C',
    'hum':'humidity_percent',
    'wind_speed':'wind_speed_kph',
    'weather_code':'weather',
    'is_holiday':'is_holiday',
    'is_weekend':'is_weekend',
    'season':'season'
}

data.rename(new_column,axis=1,inplace=True)
```

```python
data
```

```python
data['humidity_percent']=data['humidity_percent']/100
data
```

```python
season_dict = {
    '0.0':'spring',
    '1.0':'summer',
    '2.0':'autumn',
    '3.0':'winter'
}


weather_dict = {
    '1.0':'Clear',
    '2.0':'Scattered clouds',
    '3.0':'Broken clouds',
    '4.0':'Cloudy',
    '7.0':'Rain',
    '10.0':'Rain with thunderstorm',
    '26.0':'Snowfall'
}



```

```python
data.season = data.season.astype('str')
data.season = data.season.map(season_dict)
data.weather = data.weather.astype('str')
data.weather = data.weather.map(weather_dict)
data
```

```python
data.to_excel('london_bikes_data_final.xlsx',sheet_name='Bike_Data')
```
